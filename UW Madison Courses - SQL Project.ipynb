```python
# Install packages to create SQL instance in Python Jupyter
import sqlite3
import csv
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
pd.set_option('display.max_columns', None)
```

```python
# Connect to a SQLite database (this will create the database if it doesn't exist)
# Load data in as well
conn = sqlite3.connect('your_database.db')
cursor = conn.cursor()

#############
# Course Offerings
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS COURSE_OFFERINGS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
COURSE_OFFERINGS = '''
CREATE TABLE IF NOT EXISTS COURSE_OFFERINGS (
    uuid STR PRIMARY KEY,
    course_uuid STR,
    term_code INT,
    name STR
)
'''
cursor.execute(COURSE_OFFERINGS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\course_offerings.csv'
TABLE_NAME = 'COURSE_OFFERINGS'

with open(csv_filename, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (uuid, course_uuid,term_code,name) VALUES (?,?,?,?)"
                data_to_insert = (row[0], row[1], int(row[2]), row[3])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()



#############
# Courses
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS COURSES"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
COURSES = '''
CREATE TABLE IF NOT EXISTS COURSES (
    uuid STR PRIMARY KEY,
    name STR,
    number INTEGER
)
'''
cursor.execute(COURSES)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\courses.csv'
TABLE_NAME = 'COURSES'

with open(csv_filename, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (uuid, name, number) VALUES (?,?,?)"
                data_to_insert = (row[0], row[1], int(row[2]))
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()

#############
# GRADE DISTRIBUTION
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS GRADE_DISTRIBUTIONS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
GRADE_DISTRIBUTIONS = '''
CREATE TABLE IF NOT EXISTS GRADE_DISTRIBUTIONS (
    course_offering_uuid STR, 
    section_number INTEGER,
    a_count INTEGER,
    ab_count INTEGER,
    b_count INTEGER,
    bc_count INTEGER,
    c_count INTEGER,
    d_count INTEGER,
    f_count INTEGER,
    s_count INTEGER,
    u_count INTEGER,
    cr_count INTEGER,
    n_count INTEGER,
    p_count INTEGER,
    i_count INTEGER,
    nw_count INTEGER,
    nr_count INTEGER,
    other_count INTEGER
)
'''
cursor.execute(GRADE_DISTRIBUTIONS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\grade_distributions.csv'
TABLE_NAME = 'GRADE_DISTRIBUTIONS'

with open(csv_filename, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (course_offering_uuid,section_number,a_count,ab_count,b_count,bc_count,c_count,d_count,f_count,s_count,u_count,cr_count,n_count,p_count,i_count,nw_count,nr_count,other_count) VALUES (?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?,?)"
                data_to_insert = (row[0],int(row[1]),int(row[2]),int(row[3]),int(row[4]),int(row[5]),int(row[6]),int(row[7]),
                                 int(row[8]),int(row[9]),int(row[10]),int(row[11]),int(row[12]),int(row[13]),int(row[14]),
                                 int(row[15]),int(row[16]),int(row[17]))
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()



#############
#INSTRCUTORS
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS INSTRUCTORS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
INSTRUCTORS = '''
CREATE TABLE IF NOT EXISTS INSTRUCTORS (
    id INT PRIMARY KEY,
    name STR
)
'''
cursor.execute(INSTRUCTORS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\instructors.csv'
TABLE_NAME = 'INSTRUCTORS'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (id, name) VALUES (?,?)"
                data_to_insert = (int(row[0]), row[1])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()

#############
#ROOMS
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS ROOMS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
ROOMS = '''
CREATE TABLE IF NOT EXISTS ROOMS (
    uuid STR PRIMARY KEY,
    facility_code STR,
    room_code STR NULL
)
'''
cursor.execute(ROOMS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\rooms.csv'
TABLE_NAME = 'ROOMS'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (uuid, facility_code,room_code) VALUES (?,?,?)"
                data_to_insert = (row[0], row[1], row[2])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()


#############
#SCHEDULES
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS SCHEDULES"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
SCHEDULES = '''
CREATE TABLE IF NOT EXISTS SCHEDULES (
    uuid STR PRIMARY KEY,
    start_time INT ,
    end_time INT,
    mon BIT,
    tues BIT,
    wed BIT, 
    thurs BIT,
    fri BIT,
    sat BIT,
    sun BIT

)
'''
cursor.execute(SCHEDULES)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\schedules.csv'
TABLE_NAME = 'SCHEDULES'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (uuid,start_time,end_time,mon,tues,wed,thurs,fri,sat,sun) VALUES (?,?,?,?,?,?,?,?,?,?)"
                data_to_insert = (row[0], int(row[1]), int(row[2]), row[3], row[4], row[5], row[6], row[7], row[8], row[9])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()

#############
#SECTIONS
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS SECTIONS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
SECTIONS = '''
CREATE TABLE IF NOT EXISTS SECTIONS (
    uuid STR,
    course_offering_uuid STR,
    section_type STR,
    number INT,
    room_uuid STR NULL,
    schedule_uuid STR
)
'''
cursor.execute(SECTIONS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\sections.csv'
TABLE_NAME = 'SECTIONS'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (uuid,course_offering_uuid,section_type,number,room_uuid,schedule_uuid) VALUES (?,?,?,?,?,?)"
                data_to_insert = (row[0], row[1], row[2], int(row[3]), row[4], row[5])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()


#############
#SUBJECT_MEMBERSHIPS
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS SUBJECT_MEMBERSHIPS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
SUBJECT_MEMBERSHIPS = '''
CREATE TABLE IF NOT EXISTS SUBJECT_MEMBERSHIPS (
    subject_code STR,
    course_offering_uuid STR

)
'''
cursor.execute(SUBJECT_MEMBERSHIPS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\subject_memberships.csv'
TABLE_NAME = 'SUBJECT_MEMBERSHIPS'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (subject_code, course_offering_uuid) VALUES (?,?)"
                data_to_insert = (row[0], row[1])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()


#############
#SUBJECTS
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS SUBJECTS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
SUBJECTS = '''
CREATE TABLE IF NOT EXISTS SUBJECTS (
    code STR,
    name STR,
    abbreviation STR

)
'''
cursor.execute(SUBJECTS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\subjects.csv'
TABLE_NAME = 'SUBJECTS'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (code,name,abbreviation) VALUES (?,?,?)"
                data_to_insert = (row[0], row[1], row[2])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()


#############
#TEACHINGS
#############

# Drop the table if it exists
drop_table_query = "DROP TABLE IF EXISTS TEACHINGS"
cursor.execute(drop_table_query)
conn.commit()

# Then create a table in the database
TEACHINGS = '''
CREATE TABLE IF NOT EXISTS TEACHINGS (
    instructor_id INT,
    section_uuid STR PRIMARY KEY
)
'''
cursor.execute(TEACHINGS)
conn.commit()

# Read csv into SQL Table
csv_filename = r'C:\Users\michi\Downloads\UW_Madison_Data\teachings.csv'
TABLE_NAME = 'TEACHINGS'

with open(csv_filename, encoding="utf8") as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)
    
        for row in csv_reader:
                insert_data_query = "INSERT INTO "+TABLE_NAME+" (instructor_id,section_uuid) VALUES (?,?)"
                data_to_insert = (int(row[0]), row[1])
                cursor.execute(insert_data_query, data_to_insert)
conn.commit()

# # Close the cursor and connection
cursor.close()
conn.close()
```

```python
conn = sqlite3.connect('your_database.db')
cursor = conn.cursor()


select_query = '''
    SELECT 
        COURSES.name as Course_Name,
        SUBJECTS.name as Subject_Name,
        AVG(GRADE_DISTRIBUTIONS.a_count),
        AVG(GRADE_DISTRIBUTIONS.ab_count),
        AVG(GRADE_DISTRIBUTIONS.b_count),
        AVG(GRADE_DISTRIBUTIONS.bc_count),
        AVG(GRADE_DISTRIBUTIONS.c_count),
        AVG(GRADE_DISTRIBUTIONS.d_count),
        AVG(GRADE_DISTRIBUTIONS.f_count),
        AVG(GRADE_DISTRIBUTIONS.s_count),
        AVG(GRADE_DISTRIBUTIONS.u_count),
        AVG(GRADE_DISTRIBUTIONS.cr_count),
        AVG(GRADE_DISTRIBUTIONS.n_count),
        AVG(GRADE_DISTRIBUTIONS.p_count),
        AVG(GRADE_DISTRIBUTIONS.i_count),
        AVG(GRADE_DISTRIBUTIONS.nw_count),
        AVG(GRADE_DISTRIBUTIONS.nr_count),
        AVG(GRADE_DISTRIBUTIONS.other_count),
        
        
        AVG(GRADE_DISTRIBUTIONS.a_count) + AVG(GRADE_DISTRIBUTIONS.ab_count) + 
        AVG(GRADE_DISTRIBUTIONS.b_count) + AVG(GRADE_DISTRIBUTIONS.bc_count) + 
        AVG(GRADE_DISTRIBUTIONS.c_count) + AVG(GRADE_DISTRIBUTIONS.d_count) + 
        AVG(GRADE_DISTRIBUTIONS.f_count) + AVG(GRADE_DISTRIBUTIONS.s_count) +
        AVG(GRADE_DISTRIBUTIONS.u_count) + AVG(GRADE_DISTRIBUTIONS.cr_count) +
        AVG(GRADE_DISTRIBUTIONS.n_count) + AVG(GRADE_DISTRIBUTIONS.p_count) + 
        AVG(GRADE_DISTRIBUTIONS.i_count) + AVG(GRADE_DISTRIBUTIONS.nw_count) + 
        AVG(GRADE_DISTRIBUTIONS.nr_count) + AVG(GRADE_DISTRIBUTIONS.other_count) AS AVERAGE_Grade_SUM
        
    FROM 
        SECTIONS
    JOIN
        COURSE_OFFERINGS ON SECTIONS.course_offering_uuid = COURSE_OFFERINGS.uuid
    JOIN 
        COURSES ON COURSE_OFFERINGS.course_uuid = COURSES.uuid
    JOIN 
        GRADE_DISTRIBUTIONS ON SECTIONS.course_offering_uuid = GRADE_DISTRIBUTIONS.course_offering_uuid
    JOIN
        TEACHINGS ON SECTIONS.uuid = TEACHINGS.section_uuid
    JOIN
        INSTRUCTORS ON TEACHINGS.Instructor_Id = INSTRUCTORS.id
    JOIN 
        SUBJECT_MEMBERSHIPS ON SECTIONS.course_offering_uuid = SUBJECT_MEMBERSHIPS.course_offering_uuid
    JOIN 
        SUBJECTS ON SUBJECT_MEMBERSHIPS.subject_code = SUBJECTS.code
    GROUP BY 
        Course_Name
    HAVING
        AVERAGE_Grade_SUM is not 0 
    ORDER BY
        AVERAGE_Grade_SUM DESC
    LIMIT 
        10


'''
cursor.execute(select_query)

# Use Pandas to read the query result into a DataFrame
df = pd.read_sql_query(select_query, conn)


cursor.close()
conn.close()

df
```

```python

# Choose lower and upper bounds for the colormap
lower_bound_color = 'indianred'  # A darker shade of red
upper_bound_color = 'darkred'

# Create a gradient of red based on numeric values
norm = plt.Normalize(min(df['AVERAGE_Grade_SUM']), max(df['AVERAGE_Grade_SUM']))
colors = plt.cm.Reds(norm(df['AVERAGE_Grade_SUM']))

# Clip lower bound color to avoid complete white
colors[colors[:, 3] < 0.2] = plt.cm.Reds(norm(min(df['AVERAGE_Grade_SUM'])))



# Create a bar plot with color gradient
plt.bar(df['Course_Name'], df['AVERAGE_Grade_SUM'], color=colors)
plt.xlabel('Courses')
plt.ylabel('Average Class Size')
plt.title('Top 10 Largest Classes on Average')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
conn = sqlite3.connect('your_database.db')
cursor = conn.cursor()


select_query = '''
    SELECT 
        SUBJECTS.name as Subject_Name,  
        AVG(GRADE_DISTRIBUTIONS.a_count), 
        AVG(GRADE_DISTRIBUTIONS.a_count) + AVG(GRADE_DISTRIBUTIONS.ab_count) + 
        AVG(GRADE_DISTRIBUTIONS.b_count) + AVG(GRADE_DISTRIBUTIONS.bc_count) + 
        AVG(GRADE_DISTRIBUTIONS.c_count) + AVG(GRADE_DISTRIBUTIONS.d_count) + 
        AVG(GRADE_DISTRIBUTIONS.f_count) AS AVERAGE_Grade_SUM ,
        
        (AVG(GRADE_DISTRIBUTIONS.a_count) * 100.0 / (AVG(GRADE_DISTRIBUTIONS.a_count) + AVG(GRADE_DISTRIBUTIONS.ab_count) + 
        AVG(GRADE_DISTRIBUTIONS.b_count) + AVG(GRADE_DISTRIBUTIONS.bc_count) + 
        AVG(GRADE_DISTRIBUTIONS.c_count) + AVG(GRADE_DISTRIBUTIONS.d_count) + 
        AVG(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_A
        
    FROM 
        SECTIONS
    JOIN
        COURSE_OFFERINGS ON SECTIONS.course_offering_uuid = COURSE_OFFERINGS.uuid
    JOIN 
        COURSES ON COURSE_OFFERINGS.course_uuid = COURSES.uuid
    JOIN 
        GRADE_DISTRIBUTIONS ON SECTIONS.course_offering_uuid = GRADE_DISTRIBUTIONS.course_offering_uuid
    JOIN
        TEACHINGS ON SECTIONS.uuid = TEACHINGS.section_uuid
    JOIN
        INSTRUCTORS ON TEACHINGS.Instructor_Id = INSTRUCTORS.id
    JOIN 
        SUBJECT_MEMBERSHIPS ON SECTIONS.course_offering_uuid = SUBJECT_MEMBERSHIPS.course_offering_uuid
    JOIN 
        SUBJECTS ON SUBJECT_MEMBERSHIPS.subject_code = SUBJECTS.code
    GROUP BY 
        Subject_Name
    HAVING 
        AVERAGE_Grade_SUM > 10 
    ORDER BY
        percentage_grade_A DESC
        

'''
cursor.execute(select_query)

# Use Pandas to read the query result into a DataFrame
df = pd.read_sql_query(select_query, conn)


cursor.close()
conn.close()

df
```

```python
conn = sqlite3.connect('your_database.db')
cursor = conn.cursor()


select_query = '''
    SELECT 
     
        
        AVG(GRADE_DISTRIBUTIONS.a_count) + AVG(GRADE_DISTRIBUTIONS.ab_count) + 
        AVG(GRADE_DISTRIBUTIONS.b_count) + AVG(GRADE_DISTRIBUTIONS.bc_count) + 
        AVG(GRADE_DISTRIBUTIONS.c_count) + AVG(GRADE_DISTRIBUTIONS.d_count) + 
        AVG(GRADE_DISTRIBUTIONS.f_count) AS AVERAGE_Grade_SUM ,
        
        (SUM(GRADE_DISTRIBUTIONS.a_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_a, 
        
        (SUM(GRADE_DISTRIBUTIONS.ab_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_ab,
        
        (SUM(GRADE_DISTRIBUTIONS.b_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_b,
        
        (SUM(GRADE_DISTRIBUTIONS.bc_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_bc,
        
        (SUM(GRADE_DISTRIBUTIONS.c_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_c, 
        
        (SUM(GRADE_DISTRIBUTIONS.d_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_d,     
        
        (SUM(GRADE_DISTRIBUTIONS.f_count) * 100.0 / (SUM(GRADE_DISTRIBUTIONS.a_count) + SUM(GRADE_DISTRIBUTIONS.ab_count) + 
        SUM(GRADE_DISTRIBUTIONS.b_count) + SUM(GRADE_DISTRIBUTIONS.bc_count) + 
        SUM(GRADE_DISTRIBUTIONS.c_count) + SUM(GRADE_DISTRIBUTIONS.d_count) + 
        SUM(GRADE_DISTRIBUTIONS.f_count))) AS percentage_grade_f
        
    FROM 
        SECTIONS
    JOIN 
        GRADE_DISTRIBUTIONS ON SECTIONS.course_offering_uuid = GRADE_DISTRIBUTIONS.course_offering_uuid
    GROUP BY 
        SECTIONS.course_offering_uuid 
    HAVING 
        AVERAGE_Grade_SUM > 10 
        

'''
cursor.execute(select_query)

# Use Pandas to read the query result into a DataFrame
df = pd.read_sql_query(select_query, conn)


cursor.close()
conn.close()

df
```
